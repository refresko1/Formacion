```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(8,6))
hours=[2,3,4,5,6,7,8]
exam_score1= [55,60,68,70,75,80,85]
exam_score2= [45,70,78,80,85,86,89]

plt.plot(hours, exam_score1,marker='o', color='g', linestyle='-',linewidth=2, label='Estu1')
plt.plot(hours, exam_score2,marker='s', color='b', linestyle='--', linewidth=2,label='Estu2')
plt.title("dispersion del horas de estudio en 2 estudiantes")
plt.xlabel("horas")
plt.ylabel('Puntaje examen')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python

# Generación de datos
dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales = np.random.randint(1000, 5000, size=12)

sales_data = pd.DataFrame({'Dates': dates, 'Sales': sales})

# Graficar
plt.plot(sales_data['Dates'], sales_data['Sales'], marker='o', linestyle='-', label='Ventas mensuales')

# Configuración del formato de fechas en el eje X
ax = plt.gca()  # Obtén el eje actual
ax.xaxis.set_major_formatter(DateFormatter('%b %Y'))  # Usa el método correcto

# Personalización de la gráfica
plt.xticks(rotation=45)
plt.title('Análisis de ventas mensuales')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt2
import numpy as np
```

```python
x = np.linspace(10,20,100)
y = np.sin(x)

data = np.random.randn(100)
```

```python
gs = gridspec.GridSpec(2,2, height_ratios=[2,1],width_ratios=[1,1])
fig = plt2.figure(figsize =(10,8))

#1er Subplot ocupara toda la primera fila
ax1 =fig.add_subplot(gs[0,:])
ax1.plot(x,y, color = 'blue')
ax1.set_title('graficacion de seno de x')
ax1.set_xlabel('X')
ax1.set_ylabel('Sin(x)')

#2do Subplot ocupara segunda fila, primera colum
ax2 =fig.add_subplot(gs[1,0])
ax2.hist(data,bins=20, color = 'purple',edgecolor='black')
ax2.set_title('Histo')
ax2.set_xlabel('valueX')
ax2.set_ylabel('Frecuencia')

#3er Subplot ocupara segunda fila, Segunda colum
ax3 =fig.add_subplot(gs[1,1])
ax3.scatter(x,y, color = 'red')
ax3.set_title('Dispersion de Seno')
ax2.set_xlabel('valueX')
ax3.set_ylabel('Frecuencia')

plt2.tight_layout()
plt2.show()
```
