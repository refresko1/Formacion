```python
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x='tip',bins=25,cumulative=False,hue='sex',stat='count',multiple='dodge')
plt.show()
```

```python

sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False,fill=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='day',kind='hist',multiple='stack')
plt.show()
```
