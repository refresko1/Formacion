```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country=['India','Japon','Mexico','Colombia', 'Germany']
pop = [1000,1500,800,1850,900]
plt.bar(country,pop,width=0.5,color=['red','blue'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia', 'Alemania'),rotation=15)
plt.title("poblacion de paises")
plt.xlabel("pais")
plt.ylabel('Poblacion en miles de millones')
plt.show()
```

```python
country=['India','Japon','Mexico','Colombia', 'Germany']
pop = [1000,1500,800,1850,900]
plt.barh(country,pop,color=['red','blue'])
plt.yticks(np.arange(5),('India','Japon','Mexico','Colombia', 'Alemania'),rotation=15)
plt.title("poblacion de paises")
plt.ylabel("pais")
plt.xlabel('Poblacion en miles de millones')
plt.show()
```
