```python
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip', hue='smoker',kind='hist',marginal_ticks=True,marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
#plt.legend(loc='upper left',bbox_to_anchor=(1.12,0.6))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex',palette='dark',corner=True,kind='scatter')
plt.show()
```
