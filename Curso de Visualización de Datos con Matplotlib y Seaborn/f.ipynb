```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
fig,ax =plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'b--')
ax.plot(x,x+2,'r-')
ax.plot(x,x+3,'yo-')
ax.plot(x,x+4,'go:')
```

```python
plt.style.use('dark_background')
fig,ax =plt.subplots(figsize=(8,8))
ax.plot(x,x+1, color='blue',alpha=0.8, linewidth=1, linestyle='-',marker='o',markersize=8, markerfacecolor='orange')
ax.plot(x,x+2, color="red",linewidth=2, linestyle='--',marker='x',markersize=8, markerfacecolor='orange')
ax.plot(x,x+3, color='#66FFCC',linewidth=2, linestyle='dashed',marker='v',markersize=8, markerfacecolor='orange')
ax.plot(x,x+4, color='yellow',linewidth=1, linestyle=':',marker='P',markersize=8, markerfacecolor='orange')
```
