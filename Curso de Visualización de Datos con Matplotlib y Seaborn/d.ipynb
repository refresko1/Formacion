```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,1].plot(x,np.sin(x),'r')
axes[0,2].plot(x,np.tan(x),'g')
axes[0,3].plot(x,np.cos(x)**2,'y')

# ax2.plot(y,x,'r')
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,1].plot(x,np.sin(x),'r')
axes[0,2].plot(x,np.tan(x),'g')
axes[0,3].plot(x,np.cos(x)**2,'y')

axes[1,0].plot(x,np.tanh(x),'b')
axes[1,1].plot(x,np.sin(x)/0.4,'r')
axes[1,2].plot(x,np.tan(x)*1.2,'g')
axes[1,3].plot(x,np.arcsin(x)**2,'y')
```
