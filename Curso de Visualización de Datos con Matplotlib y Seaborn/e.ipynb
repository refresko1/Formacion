```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig,axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y,'b',label="$sin(x)$")
axes[0].set_title("relacion X-Y")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend()

axes[1].plot(y,x,'b',label="$sin(y)$")
axes[1].set_title("relacion Y-X")
axes[1].set_xlabel("Y")
axes[1].set_ylabel("X")
axes[1].legend()

fig.tight_layout()
plt.show()



```

```python
plt.plot(x,y,'r', label="$sin(x)$")
plt.title('Titulo ')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='lower left', bbox_to_anchor=(0.0,0.1))
```
